```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("2023_01_sports/8_5_output.csv")
```

```python
pivot_table = df.pivot_table(
    values='mean_test_score', 
    index=['param_min_samples', 'param_alpha'],
    columns=["param_min_cluster_size", 'param_cluster_selection_method']
)
```

```python
heatmap = sns.clustermap(pivot_table, cmap='YlGnBu', annot=True, figsize=(12, 10), annot_kws={'size': 10})
heatmap.ax_heatmap.set_xlabel('min_cluster_size & cluster_selection_method', fontsize=12)  # 修改 x 轴标签
heatmap.ax_heatmap.set_ylabel('min_samples & alpha', fontsize=12)  # 修改 y 轴标签

# 设置标题
plt.title('Cluster Map of mean_test_score', fontsize=16, loc='center', pad=20)
```
